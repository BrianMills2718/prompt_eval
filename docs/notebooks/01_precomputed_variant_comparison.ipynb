# Precomputed Variant Comparison

## Purpose

Define the smallest `prompt_eval` contract needed to compare frozen outputs from external systems like `grounded-research` without reviving multiple runtime modes.

## Contract

- input: precomputed outputs keyed by `variant_name`, `input_id`, optional `replicate`
- processing: run existing trial and corpus evaluators over those outputs
- output: normal `EvalResult`
- comparison: use existing `compare_variants()` including `paired_by_input`

## First Use Case

- variant A: Tyler-literal grounded-research outputs
- variant B: calibrated-legacy grounded-research outputs
- frozen shared case set: same benchmark inputs, same saved artifacts
- no second runtime inside `grounded-research`

## Acceptance

1. `prompt_eval` can score frozen outputs directly.
2. The result is a standard `EvalResult`.
3. Paired comparison works without special-case code.
4. Observability, if enabled, is truthful about precomputed provenance.
